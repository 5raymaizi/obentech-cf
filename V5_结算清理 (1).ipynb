```python
# 清理剩余的一些套利仓位
target_env_list = ['bn_v5_pm_pro5']

# [22, 23, 24, 26]
# [14, 19, 22, 23, 25, 26, 31]
# target_env = total_env_list
target_env = target_env_list
not_symbol_list = ['BNBUSDT']
is_true = False
is_true = True
clean_usdt_value_threshold = 100

symbol_dict = binanceController.tt_symbol_dict
for env, symbol_list in symbol_dict.items():
    if env in target_env:
        querystring = {"env":env}
        print(env)
        for symbol in symbol_list:
            if symbol not in not_symbol_list:
                try:
                    print(symbol)
                    url = f"https://{binanceController.prefix}.digifinex.org/api/cf_arbitrage/{symbol}"
                    response = binanceController.verified_session.request("GET", url, cookies = binanceController.verified_cookies, params=querystring)
                    response_dict = response.json()

                    account_info = response_dict['data']['detail']['account']
        #             print(account_info)
                    baseAmount = float(account_info['baseAmount'])
                    baseBorrowed = float(account_info['baseBorrowed'])
                    poss = pd.DataFrame(response_dict['data']['detail']['pos_list']['poss'])
                    
                    posAmount = float(response_dict['data']['detail']['pos_list']['pos_amount'])
                    localPosAmount = float(response_dict['data']['detail']['pos_list']['local_pos_amount'])
                    localPosAmountMt = float(response_dict['data']['detail']['pos_list']['local_pos_amount_mt'])
                    localPosAmountTm = float(response_dict['data']['detail']['pos_list']['local_pos_amount_tm'])
                    localPosAmount = localPosAmount+localPosAmountMt+localPosAmountTm

                    

                    
                    open_allowed = bool(response_dict['data']['config']['open_allowed'])

                    if 'USDT' in symbol:
                        request_symbol_name = symbol
                    else:
                        request_symbol_name = symbol + 'T'
                    url = f'https://api.binance.com/api/v3/aggTrades?symbol={request_symbol_name}&limit=1000'
                    aggTradesDf = pd.DataFrame(requests.get(url, proxies=proxies).json())
                    cur_spot_price = float(aggTradesDf.iloc[-1]['p'])
                    if (baseAmount*cur_spot_price <= clean_usdt_value_threshold) & (posAmount>0) & (((posAmount-localPosAmount)*cur_spot_price)<clean_usdt_value_threshold) & (not open_allowed):
                        print(symbol, baseAmount, cur_spot_price, baseAmount*cur_spot_price)
                        if is_true:
                            headers = {
                                "content-type": "application/json"
                            }

                            # 平合约
                            swapOrder ={"type":"swap","side":'closeShort',"amount":str(posAmount)}
                            url = f"https://{binanceController.prefix}.digifinex.org/api/cf_arbitrage/{symbol}/addOrder?env={env}"
                            print(url, swapOrder)
                            response = binanceController.verified_session.request("POST", url,
                                                                              cookies = binanceController.verified_cookies,
                                                                              headers = headers, 
                                                                              data=json.dumps(swapOrder))
                            print(response.status_code)
                            print(response.json())  
                            time.sleep(5)

                            # 重启并关闭
                            update_json = {"change":[{"control":env, "symbol":f"{symbol}", "running":False}]}
                            print(update_json)
                            binanceController.updateConfig(update_json)
                            time.sleep(1)
                            update_json = {"change":[{"control":env, "symbol":f"{symbol}", "running":True,
                                                      "config":{"open_allowed":True, "close_allowed":True}}]}
                            print(update_json)
                            binanceController.updateConfig(update_json)
                            update_json = {"change":[{"control":env, "symbol":f"{symbol}", "running":False}]}
                            print(update_json)
                            binanceController.updateConfig(update_json)
                            time.sleep(2)

                            # 清理环境
                            clear_url = f"https://{binanceController.prefix}.digifinex.org/api/cf_arbitrage/{symbol}/clear?env={env}"
                            response = binanceController.verified_session.request("PUT", clear_url,
                                                                                cookies = binanceController.verified_cookies,
                                                                                headers=headers)
                            print(response.status_code)
                            print(response.json())  
                            time.sleep(1)

                            # 平现货
                            spotOrder ={"type":"spot","side":"sell","amount":str(baseAmount)}
                            url = f"https://{binanceController.prefix}.digifinex.org/api/cf_arbitrage/{symbol}/addOrder?env={env}"
                            print(url, spotOrder)
                            response = binanceController.verified_session.request("POST", url,
                                                             cookies = binanceController.verified_cookies,
                                                             headers = headers, 
                                                             data=json.dumps(spotOrder))
                            print(response.status_code)
                            print(response.json())    
                            time.sleep(1)

                            # 重启再关闭
                            update_json = {"change":[{"control":env, "symbol":f"{symbol}", "running":False}]}
                            print(update_json)
                            binanceController.updateConfig(update_json)
                            time.sleep(1)
                            update_json = {"change":[{"control":env, "symbol":f"{symbol}", "running":True,
                                                      "config":{"open_allowed":True, "close_allowed":True}}]}
                            print(update_json)
                            binanceController.updateConfig(update_json)
                            time.sleep(1)
                            update_json = {"change":[{"control":env, "symbol":f"{symbol}", "running":False}]}
                            print(update_json)
                            binanceController.updateConfig(update_json)
                            time.sleep(1)

                except Exception as e:
#                     print(e)
                    continue
```
